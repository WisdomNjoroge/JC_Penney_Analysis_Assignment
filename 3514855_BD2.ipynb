```python
# inspecting sample reviews by score to check for scale consistency.
for s in sorted(reviews["score"].unique()):
    print(f"\n== Reviews with score {s} ===")
    print(reviews[reviews["score"] == s]["review"].head(3).to_list())
```

```python
# reversing the score scale so that high number mean higher satisfaction
reviews["score"] = 6 - reviews["score"]

# confirming the fix worked
for s in sorted(reviews["score"].unique()):
    print(f"\n== Reviews with corrected score {s} ===")
    print(reviews[reviews["score"] == s]["review"].head(3).to_list())
```

```python
# keeping only scores between 1 and 5 for consistency
reviews = reviews[(reviews["score"] >= 1) & (reviews["score"] <= 5)]

print("Unique scores after cleaning:", reviews["score"].unique())
```

```python
# confirming the fix worked
for s in sorted(reviews["score"].unique()):
    print(f"\n== Reviews with corrected score {s} ===")
    print(reviews[reviews["score"] == s]["review"].head(3).to_list())
```

After inspecting the text of customer reviews, I found that the numerical ratings did not always match the sentiment of the written feedback. While many low scores originally reflected positive comments, there were also mixed and inconsistent examples. This suggests that reviewers may have used the ratings scale differently, or that some reviews were duplicated.

I reversed the overall scale so that the higher rating represents greater satisfaction that is 1 = Poor while 5 = Excellent. Although some ambiguity still remains in mid-range reviews, this correction provides a clearer and more interpretable analysis.

```python
print("Reviews:", list(reviews.columns))
print("Products:", list(products.columns))
print("Users:", list(users.columns))
print("Reviewers:", list(reviewers.columns))
```

Adding new columns which will make it easier to understand how satisfaction varies across different reviewers and review behaviour.

```python
# Add review length to measure how detailed each review is
reviews.loc[:, "review_length"] = reviews["review"].astype(str).str.len()

# Convert Date of Birth to proper date format
users["dob"] = pd.to_datetime(users["dob"], format="%d.%m.%Y",errors="coerce")
reviewers["dob"] = pd.to_datetime(reviewers["dob"], format="%d.%m.%Y",errors="coerce")

# Work out the age
today = pd.Timestamp("2025-01-01")
users["age"] = today.year - users["dob"].dt.year
reviewers["age"] = today.year - reviewers["dob"].dt.year

# Group users by generation
bins = [0, 25, 40, 55, 70, 100]
labels = ["Gen Z", "Millenials", "Gen X", "Bommer", "Silent Gen"]
users["age_group"] = pd.cut(users["age"], bins=bins, labels=labels)
reviewers["age_group"] = pd.cut(reviewers["age"], bins=bins, labels=labels)

print("Extra columns successfully added!")
```

## C. Data Preparation and Merging.

Now that the data looks clean and consistent, the next step is to bring all the pieces together. Each dataset holds part of the sotry - products, users, reviews and reviewer's information and so merging the datasets will help me see the bigger prcture of what drives customer satisfaction.

```python
# Merge reviews with product details using the product ID which is the uniq_id
merged = reviews.merge(products, on="uniq_id", how="left")

# Add user information using the username column
merged = merged.merge(users, on="username", how="left")

print("Data merged successfully")
print("Merged shape:", merged.shape)
display(merged.head(3))
```

```python
# check for duplicated information
duplicate_rows = merged.duplicated().sum()
print(f"Duplicate rows: {duplicate_rows}")
```

```python
# check for missing values after merging
print("\nMissing values summary:")
print(merged.isnull().sum().sort_values(ascending=False).head(10))
```

```python
# fill misiing descriptions and reviews with placeholders
merged["description"] = merged["description"].fillna("No description available")
merged["review"] = merged["review"].fillna("No review text provided")
merged["price"] = merged["price"].fillna(merged["price"].mean())

print("Missing text values filled where necessary")
```

```python
print("Final merged columns:")
print(list(merged.columns))
```

```python
print("\nPreview of final dataset:")
display(merged.sample(5))
```

Summary of Data Preparation and Merging.

This section brought all the cleaned datasets together into one complete table. I started by merging the Reviews and Products data using the product ID (uniq_id), then added the Users information using the username column.

After merging, I checked for missing values and duplicate rows to make sure the process worked well. After running the code, some product descriptions were missing, so I filled those with simple placeholders to keep the dataset consistent.

The result is a single, unified dataset that links each product review to both the reviewer's personal; details and the product's information and with this data I will now use in the next section to explore patterns, visualise trends and understand what drives customer satisfaction at JC Penney.

## D. Exploratory Analysis.

Now that the data is clean and merged, I can begin exploring it to understand what drives customer satisfaction at JC Penney. This section uses descriptive statistics, visualisations and models trends in ratings, engagement and reviewer behavior.
The main goals in this section are :
- See how customers rate products overall.
- Identify which product categories receive the highest and lowest ratings.
- Explore whether longer reviews relate to higher satisfaction.

```python
# checking the range of scores
merged["score"].value_counts().sort_index()
```

Before plotting, I checked whether the rating scale starts at 0 or 1. Since it begins at 0, it likely means 0 = 1 and 4 = 5 stars. To make the chart easier to read, I shifted the scale to +1 so that the ratings run from 1 to 5.

```python
print(sorted(merged["score"].unique()))

sns.countplot(x="score", data=merged, hue="score", palette="crest", legend=False)
plt.title("Distribution of customer ratings (1 = Poor, 5 = Excellent)")
plt.xlabel("Rating")
plt.ylabel("Number of Reviews")

#save chart
plt.show()
```

After adjusting the scale, ratings now range from 1 to 5, which aligns with a standard customer satisfaction system.
Most reviews fall between 3 and 5 stars, suggesting generally positive customer experiencesand the very low ratings are less common. The corrected scale now shows a logical pattern where higher scores reflect greater satisfaction.

```python
# calculate average rating and review count per product
product_stats = merged.groupby("name").agg(
    avg_rating=("score", "mean"),
    review_count=("score", "count")
).reset_index()

# filter products with at least 10 reviews
popular_products = product_stats[product_stats["review_count"] >= 10]

# sort and select top 10 products
top_products = popular_products.sort_values("avg_rating", ascending=False).head(10)

print("Top products created successfully")
display(top_products.head())
```

```python
import textwrap

top_products["name_wrapped"] = top_products["name"].apply(
    lambda x: "\n".join(textwrap.wrap(x, width=25))
)

# creating the plot for the top 10 products by Average Rating
plt.figure(figsize=(9,8))
sns.barplot(y="name_wrapped", x="avg_rating", data=top_products, color="teal")

for i, row in top_products.iterrows():
    plt.text(row.avg_rating - 0.15, i, f"{int(row.review_count)}",
             va="center", ha="right", color="white", fontsize=9, fontweight="bold")

plt.title("Top 10 Products by Average Rating", fontsize=13, pad=10)
plt.xlabel("Average Rating", fontsize=11)
plt.ylabel("Product Name", fontsize=11)
plt.subplots_adjust(left=0.45, right=0.95, top=0.9, bottom=0.1)
plt.show()
```

These are the highest-rated products with at least 10 reviews. All maintain near-perfect averages, which suggests consistently good customer experiences. By showing the review counts it helps confirm that these results are based on a meaningful number of opinions.

```python
# finding 10 most reviewed products
top_reviewed = merged["name"].value_counts().head(10).reset_index()
top_reviewed.columns = ["name", "review_count"]

top_reviewed["name_wrapped"] = top_reviewed["name"].apply(
    lambda x: "\n".join(textwrap.wrap(x, 25))
)

plt.figure(figsize=(10,8))
sns.barplot(x="review_count", y="name_wrapped", data=top_reviewed, color="steelblue")
plt.title("Top 10 Most Reviewed Products")
plt.xlabel("Number of Reviews")
plt.ylabel("Product Name")
plt.subplots_adjust(left=0.4, right=0.95, top=0.9, bottom=0.1)
plt.show()
```

The plot above highlights the most reviewed products. Comparing this with the previous chart shows that the most popular items are not always the ones with the highest average rating.

```python
# showing the relationship between review length and rating.
plt.figure(figsize=(8, 5))
sns.regplot(x="review_length", y="score", data=merged,
            scatter_kws={"alpha":0.4}, line_kws={"color":"red"})
plt.title("Relationship between Review Length and Rating")
plt.xlabel("Review length (characters)")
plt.ylabel("Rating (1-5)")
plt.show()
```

Although I expected longer reviews to have higher ratings, the regression shows a slightly negative trend. This suggests that customers who write very long reports are often explaining issues or complaints rather than expressing strong satisfaction. In contrast, happy customers tend to leave shorter positive comments such as "Good fit" or "Perfect quality".
The pattern is not very strong , but it highlights that detailed feedback can be useful for identifying areas of improvement, since longer reviews usually contain more constructive criticism.

```python
# plotting the average rating by age group
plt.figure(figsize=(8,5))
ax = sns.barplot(y="age_group", x="score", data=merged,
                 hue="age_group", palette="coolwarm", errorbar=None, legend=False)

for p in ax.patches:
    plt.text(p.get_width() + 0.01,
             p.get_y() + p.get_height()/2,
             f"{p.get_width():.2f}",
             va="center", fontsize=9)
             
plt.title("Average Rating by Age Group")
plt.xlabel("Average Rating")
plt.ylabel("Age Group")

plt.xlim(3.6, 4.0)
plt.tight_layout()
plt.show()
```

The average rating by age group shows remarkable consistency across generations. Gen Z, Boomers and the Silent Gen each report an identical average of 3.92 while millenials (3.91) and Gen X (3.89). This uniformity could reflect JC Penney's mid-market positioning-its products resonate equally well across generations, reducing the generational satisfaction gap often seen in fashion retail.

```python
# plotting the reviews by age group

reviews_by_age = merged["age_group"].value_counts().reset_index()
reviews_by_age.columns = ["age_group", "review_count"]
reviews_by_age = reviews_by_age.sort_values("review_count", ascending=True)

plt.figure(figsize=(8,5))
ax = sns.barplot(x="review_count", y="age_group", data=reviews_by_age, color="teal")

for i, v in enumerate(reviews_by_age["review_count"]):
    plt.text(v - 100, i, str(v), va="center", ha="right", color="white", fontsize=10)

plt.title("Number of Reviews by Age Group")
plt.xlabel("Number of Reviews")
plt.ylabel("Age Group")
plt.tight_layout()
plt.show()
```

The visualisation above highlights how review activity varies across age groups. While all generations participate in online reviews, the Boomer and Gen X collectively contribute the majority of feedback. This suggests JC Penney's strongest engagement remains among older, established shoppers who have a longer relationship with the brand.

In contrast, Millenials and especially Gen Z show lower review volumes. This could indicate a potential growth opportunity for JC Penney to build stronger digital shoppers through targeted online campaings, mobile experiences, and influencer engagement.

```python
# top 10 states by number of reviews

top_states = merged["state"].value_counts().head(10).reset_index()
top_states.columns = ["state", "review_count"]

plt.figure(figsize=(9,5))
sns.barplot(x="review_count", y="state", data=top_states,
            hue="state", palette="viridis", legend=False)
             
plt.title("Top 10 States by Number of Reviews", fontsize=13, weight="bold")
plt.xlabel("Number of Reviews")
plt.ylabel("State")

for index, value in enumerate(top_states["review_count"]):
    plt.text(value - 40, index, str(value), va="center",
             ha="right", color="white", fontsize=9, fontweight="bold")
plt.xlim(0, top_states["review_count"].max() + 100)

plt.tight_layout()
plt.show()
```

The chart shows that states such as Vermont, Delaware and Massachusetts produce the most reviews, each contributing over 600 entries. The spread is fairly even among the top ten, indicating that JC Penney maintains consistent customer engagement across dverse U.S regions rather than being dependent on just one major market.
